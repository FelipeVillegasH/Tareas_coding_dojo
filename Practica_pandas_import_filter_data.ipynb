```python
# definimos la variable de filtro
filteruptojuly= df['Week of']> '2020-06-29'
# tome en cuenta que se deben poner los corchetes excluyendo la condicion
# se debe tomar en cuenta que no se puede escribir la fecha 2020-06-31 requerida en la inficacion del ejercicio porque nos devolver'a error 
# al no existir dicha fecha en la base de datos, por lo que se debe procurar buscar en la base la fecha mas cercana y tomarla como base de referencia 
# por esta razon ponemos > y no >=
filteruptojuly
# esto imprime los resultados en True or False (booleans)
# para imprimir la tabla debemos asignar una variable que utilice esta condicion para reescribir la tabla
# cabe recalcar el valor atipico al inicio de la tabla con fecha de diciembre y por lo tanto, sale True
```

```python
# aplicamos variable de filtro en un nuevo dataframe
df_upto_july= df[filteruptojuly]
df_upto_july
# una vez generada la variable que nos imprime la nueva tabla filtrada procedemos a copiar el nombre en la linea subsiguiente para imprimirla y revisar 
```

```python
# un vez mas definimos la varible que contiene el filtro a usar 
change1 = df_upto_july['Change 2019-2020'] > 1  
# NOTA : no poner los integers entre comillas, distinguir entre strings, floats, integers, and so on 
# ya que de ponerlos entre comillas asi: change1 = df_upto_july['Change 2019-2020'] > '1'  nos devolvera error
```
