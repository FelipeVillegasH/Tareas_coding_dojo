Description: hereby we use filters to replace the NaN values in the height variable to then obtain the average height of the athlets by year by using .groupby

```python
import pandas as pd
from google.colab import drive 
drive.mount('/content/drive')
```

```python
athletdf=pd.read_csv('/content/drive/MyDrive/CodingDojo/tareas/core/Semana 3 Graphics/athleteEventsNoPersonal.csv')
```

```python
athletdf.head()
```

```python
athletdf.info()
```

```python
athletdf['Height']=athletdf['Height'].fillna(0)
athletdf.info()
```

```python
Average_AthleteHeight_by_year= athletdf.groupby(['ID','Year'])[['Height']].mean()
Average_AthleteHeight_by_year
Average_AthleteHeight_by_year=Average_AthleteHeight_by_year.sort_values(by=['Year'],ascending=[False])
Average_AthleteHeight_by_year
# the cero values correspond to the non registered height values
```
