```python
# Cantidad de sobrevivientes 
df_Titanic_survivals= df_Titanic['Survived'].sum(['Survived']==1)
df_Titanic_survivals
```

```python
# porcentaje de sobrevivientes
Percentage_of_Survival= df_Titanic_survivals/891
Percentage_of_Survival
# 38.38%
```

```python
# data frame con inde3xes de valores booleans de sobrevivientes
df_Survivals= df_Titanic['Survived']==1
df_Survivals
```

```python
# aplicamos un filtro de sobrvivientes por sexo masculino, con esto configuramos un condicional logico para imprimir los sobrevivientes hombres
df_Titanic_Males=df_Titanic['Sex']=='male'
df_survivals_by_sex = df_Titanic[(df_Survivals)&(df_Titanic_Males)]
df_survivals_by_sex
```

```python
# tambien podemos obtener la suma de sobrevivientes hombres con el metodo aggregate sum
Male_survivals= df_survivals_by_sex['Survived'].sum(['Survived']==1)
Male_survivals
```

```python
# por exclusion obtenemos el numero de mujeres sobrevivientes
Female_survivals=df_Titanic_survivals-Male_survivals
Female_survivals
```

```python
# Posteriormentre obtenemos el porcentaje dividiendo para el total de non-null values
Percentage_Female_survivals=Female_survivals/891
Percentage_Male_survivals=Male_survivals/891
Percentage_Female_survivals
# Porcentaje de mujeres sobrevivientes: 26.15%
```

```python
Percentage_Male_survivals
# hombres sobrevivientes: 12.23%
```

```python
# Usando Filtros para male 
# generamos las variables condicionales del filtrado a realizar: sobrevivientes- hombres
df_Survivals_Filter=df_Titanic['Survived']== 1
df_Sex_filter=df_Titanic['Sex']== 'male'
# con el metodo .loc, fijamos las condiciones y creamos una nueva variable que solo contenga los datos de aquellos que sobrevivieron y fueron hombres 
df_filter_Male_Survivals= df_Titanic.loc[df_Survivals_Filter & df_Sex_filter,:]
# comprbamos que se aplico bien el filtro con .info()
df_filter_Male_Survivals.info()
df_filter_Male_Survivals.sum()
# en efecto lo sobrevivientes hombres fueron 109 lo que comprueba los metodos anteriores 
```

```python
# tambien podemos realizar esto con el metodo groupby apalancados en el metodo aggregate suma 
df_groupbyed=df_Titanic.groupby(['Fare','Survived']).sum()
df_groupbyed
```

```python
# fijamos el filtro de aquellos que pagaron menos de 10  y utilizando el filtro anterior de sobrevivientes obtenemos aquellos sobrevivientes, hombres o mujeres
# que pagaron menos de 10 
df_Fare_Filter=df_Titanic['Fare']<10
df_filter_Fare_Survivals= df_Titanic.loc[df_Survivals_Filter & df_Fare_Filter,:]
# sumamos el resultado y obtenemos
df_filter_Fare_Survivals.sum()
# 67 hombres y mujeres que sobrevivieron y pagaron menos de 10
Survivlas_Fare_percentage= 67/891*100
Survivlas_Fare_percentage
# que corresponden a un porcentaje de 7.52%
```

```python
# usando Groupby 
# Groupby funciona : (['agrupas observaciones'])[['agrupas variables']]. segun esta condicion operar las observaciones de las variables
df_Titanic.groupby(['Survived'])[['Age']].mean()
# obtenemos la edad promedio de sobrevivientes 1 y no sobrevivientes 0
```

```python
df_Titanic.groupby(['Survived','Sex'])[['Age']].mean()
# por ultimo agrupamos sobrevivientes por sexo y obtenemos la edad promedio 
# es asi como la edad promedio de mujeres sobrevivientes era de 28 years mientras que la de los hombres sobrevivientes de 27 years 
# por otro lado la edad promedio de mujeres no sobrevivientes era de 25 years y la de los hombres era de 32
```
