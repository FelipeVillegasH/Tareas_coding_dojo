```python
# Reemplazamos los nombres de las columnas con el metodo de diccionario
df= df.rename(columns={'0':'ID_Flujo',
              '1':'Fecha',
              '2':'Hora',
              '3':'Conteo'})
```

```python
# revisamos los cambios con el metodo .head
df.head()
```

```python
df.info()
# se identifica que la fecha esta en formato object, por lo que necesitamos cambiarlo a formato datetime64, 
# probablemente la hora necesite estar en el mismo formato tambien (objeto hace referencia a que se esta trabajando con strings)
```

```python
# para convertir datos a tipo numerico:
# df[cols]=df[cols].apply(pd.to_numeric,errors='coerce')
# para convertir de datos numericos a datetime64:
# df['Fecha']=pd.to_datetime(df['Fecha'])
# seleccionamos dataframe ['la columna a cambiar']=pd. argumento de cambio a fecha (dataframe['columna a cambiar'])
df['Fecha']=pd.to_datetime(df['Fecha'])
df['Hora']=pd.to_datetime(df['Hora'])
df.info()
```

```python
# generamos las variables condicionales del filtrado a realizar, en este caso queremos seleccionar una fecha especifica solo flujos de entrada
df_Fecha_Filter=df['Fecha']== '07/24/05'
df_Flujo_filter=df['ID_Flujo']==7
# con el metodo .loc, fijamos las condiciones y creamos una nueva variable que solo contenga los datos del 07/24/05 de flujos de entrada
df_filter1= df.loc[df_Fecha_Filter & df_Flujo_filter,:]
# comprbamos que se aplico bien el filtro con .info()
df_filter1.info()
df_filter1
```

```python
# ordenamos los valores filtrados de mayor a menor (descendente)
df_filter1=df_filter1.sort_values(by=['Conteo'],ascending=[False])
# El True en ascending=False define el contrario a Descendiente, si quisieramos en orden ascendiente escribimos la condicion boolean True
```

```python
# generamos la variable de los 10 primeros valores correspondientes a la mayor cantidad de entradas el 24/07/2005 por hora
Top_10=df_filter1.head(10)
Top_10
```

```python
min=Top_10.min()
min
```

```python
# generamos una variable, nombramos la variable a modificar en este caso Top_10 con el punto nombramos la columna y con el apply aplicamos la resta 
# posteriormente, como esta 'ultima variable de Top_restado solo trabaja y por lo tanto, solo ilustra la columna de conteo, no necesitamos nombrarla
# para sacar la media de los datos

Top_restado= Top_10.Conteo.apply(lambda x: x-1)
Top_restado
```

```python
Top_restado.mean()
```
